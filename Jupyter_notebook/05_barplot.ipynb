```python
#-Example 1. (Flower)
#-import libraries.
import seaborn as sns
import matplotlib.pyplot as plt

#-load Dataset.
phool = sns.load_dataset("iris")
phool

#-draw a barplot.
sns.barplot(x="sepal_length", y="sepal_width", data=phool)
plt.show
```

```python
#-Example 2. (Flower)
#firstly import libraries
import seaborn as sns
import matplotlib.pyplot

#Secondly load dataset
phool = sns.load_dataset("iris")
phool

#Third draw barplot
sns.barplot(x="species", y="sepal_width", data=phool)
plt.show()
```

```python
#Example 3. (titanic)
#Firstly import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Secondly load dataset
kashti =sns.load_dataset("titanic")
kashti

#Third load barplot and adding order by
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female","male"])
plt.show()
```

```python
#Example 4. (adding color)
#firstly import libraries.
import seaborn as sns
import matplotlib.pyplot as plz

#Secondly load_dataset.
kashti = sns.load_dataset("titanic")
kashti

#Third adding barplot and adding color.
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female","male"], color = "red")
plt.show()
```

```python
# Example 5. (remove bar(confidence interval))
#-firstly import library
import seaborn as sns
import matplotlib.pyplot as plt

#-secondly load_dataset
kashti = sns.load_dataset("titanic")
kashti

#Third adding barplot and removing ci(confidence interval) mean(curve upper bars)
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female","male"], color="green", ci=None)
plt.show()
```

```python
# Example 6. (changing color palette)
#-Firstly import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#-load-dataset
kashti = sns.load_dataset("titanic")
kashti

#-adding barplot with color palette
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female","male"], color="green", ci=None, palette='pastel')
plt.show()
```
